# 3-Data Cleaning

handling missing values in BasePay, OvertimePay, and OtherPay with medianto avoid sensitivity to outliers.

because the mean can be  influenced by these outliers,The median is less sensitive to outliers, making it a more strong measure of central tendency in such cases.



```python

df['BasePay'].fillna(df['BasePay'].median(), inplace=True)
df['OvertimePay'].fillna(df['OvertimePay'].median(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].median(), inplace=True)
df['Benefits'].fillna(df['Benefits'].median(), inplace=True)
```

Both Notes and Status columns have a large number of missing values. 

so dropping these columns is best solution

```python

df.drop(['Notes', 'Status'], axis=1, inplace=True)
```

```python
# List of columns to check for negative values
check_col = ['TotalPay', 'BasePay', 'Benefits', 'TotalPayBenefits', 'OvertimePay', 'OtherPay']

# Check for and delete rows with negative values 
for column in check_col:
    df = df[df[column] >= 0]

# Drop rows with the value of zero
df= df[df['TotalPay'] > 0]
```

```python
df.isnull().sum()
```

# 4-Basic Data Visualization

```python
# Create a histogram to visualize the distribution of salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=20, kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
#use pie charts to represent the proportion of employees in different departments
# Extract department information from the 'JobTitle' column
department_df = df['JobTitle'].str.extract(r'\((.*?)\)')
department = department_df[0]
department_lower=department.str.lower()
department_counts=department_lower.value_counts()
# Plot a pie chart
plt.figure(figsize=(30, 30))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

# Simple Correlation Analysis:

```python
df.info()
```

```python
# Select numerical columns for correlation analysis

numeric_df=df.drop(["Id","JobTitle","EmployeeName","Agency"], axis=1)
# Calculate correlation matrix
corr_matrix = numeric_df.corr()
# Plot the correlation matrix
plt.figure(figsize=(12, 9))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Identify any correlation between salary and another numerical column
correlation = df['TotalPay'].corr(df['BasePay'])
print(f"Correlation between Salary and BasePay: {correlation}")
```

```python
# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x='BasePay', y='TotalPay', data=df)
plt.title('Scatter Plot: Salary vs BasePay')
plt.xlabel('BasePay')
plt.ylabel('Salary')
plt.show()
```
